## üóÇÔ∏è Simplifying Book Categories for Recommendation

The original dataset contains a wide range of **specific categories**, which can make it difficult to group books meaningfully for recommendation or analysis. To address this, we define a **category mapping dictionary** to group similar genres into broader, simplified categories.

```python
books["simple_categories"] = books["categories"].map(category_mapping)
```

## üß† Zero-Shot Book Category Classification using Hugging Face

We evaluate how well a Hugging Face zero-shot classifier can distinguish between **Fiction** and **Nonfiction** book descriptions using natural language inference.

---

### ‚öôÔ∏è Step 1: Define Prediction Function

We define a function `generate_predictions()` that uses the `pipeline()` API to get prediction scores for each category and returns the label with the **highest confidence**.

### üîÅ Step 2: Evaluate on 100 Fiction & 100 Nonfiction Samples

We test the classifier on 100 samples from each category and collect predictions

---

* Fiction samples are tested first

* Then Nonfiction samples are tested

* We store both actual and predicted labels for evaluation


```python
import numpy as np
max_index = np.argmax(pipe(sequence, fiction_categories)["scores"])
max_label = pipe(sequence, fiction_categories)["labels"][max_index]
max_label
```

```python
def generate_predictions (sequence, categories):
	predictions = pipe (sequence, categories)
	max_index = np.argmax(predictions["scores"])
	max_label = predictions["labels"][max_index]
	return max_label
```

```python
from tqdm import tqdm

actual_cats = []
predicted_cats = []
for i in tqdm(range(0,100)):
  sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[i]
  predicted_cats.append(generate_predictions(sequence, fiction_categories))
  actual_cats.append("Fiction")
```

```python
for i in tqdm(range(0,100)):
  sequence = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]
  predicted_cats.append(generate_predictions(sequence, fiction_categories))
  actual_cats.append("Nonfiction")
```

```python
predictions_df = pd.DataFrame({
	"actual_categories": actual_cats,
	"predicted_categories": predicted_cats
})
```

```python
predictions_df
```

```python
predictions_df["correct_prediction"] = np.where(predictions_df["actual_categories"] == predictions_df["predicted_categories"],1,0)
```

```python
isbns =[]
predicted_cats = []

missing_cats = books.loc[books["simple_categories"].isna(), ["isbn13","description"]].reset_index(drop=True)
```

```python
missing_cats.shape
```

```python
for i in tqdm(range(len(missing_cats))):
	sequence = missing_cats["description"][i]
	predicted_cat = generate_predictions(sequence, fiction_categories)
	predicted_cats.append(predicted_cat)
	isbns.append(missing_cats["isbn13"][i])
```

## üß™ Validating Predictions and Handling Missing Categories

After classifying 100 Fiction and 100 Nonfiction samples, we move forward to:

1. **Validate prediction accuracy**
2. **Predict missing categories**
3. **Refine genre assignments for popular fiction genres**

---

### üìà Step 1: Store and Evaluate Predictions

We store actual and predicted categories in a DataFrame and flag whether each prediction was correct.

### üîç Step 2: Predict Categories for Missing Values
We handle rows where simple_categories are missing by predicting them using our zero-shot classifier.

### üîÑ Step 3: Merge Back and Update Missing Categories
We merge the predictions into the main dataset and fill in the missing `simple_categories`

### üéØ Step 4: Isolate Popular Fiction Genres
Finally, we identify books that fall under popular fiction subgenres such as:

`Romance`, `Science Fiction`, `Fantasy`, `Mystery`, `Thriller`, `Horror`

`Comedy`, `Crime`,`Historical`




```python
missing_predictions_df = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_cats})
missing_predictions_df
```

```python
books = pd.merge(books, missing_predictions_df, on="isbn13", how="left")
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])
books = books.drop(columns=["predicted_categories"])
```
